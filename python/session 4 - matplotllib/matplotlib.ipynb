
<div dir='rtl'>برای رسم هر تابع دلخواه می‌توان از تابع <code>plot</code> استفاده کرد. این تابع با ورودی گرفتن مجموعه مقادیر <code>x</code> و <code>y</code> تابعی را با استفاده از آن‌ها رسم می‌کند.
در مثال زیر تابع <code>sin</code> در پنچاه نقطه در بازه‌ی $[-\pi, \; \pi]$ رسم شده است.</div>

````python
pi = np.pi
x = np.linspace(start=-pi, stop=pi, num=10)
y = np.sin(x)
plt.plot(x, y)
plt.show()
````

<div dir='rtl'>برای نمایش نمودار از تابع <code>show</code> استفاده می‌کنیم.</div>
<br>
<div dir='rtl'>در نمودار زیر با تغییر پارامتر‌های <code>markersize</code>، <code>marker</code>، ‍<code>linewidth</code>، <code>linestyle</code> و <code>color</code> صورت‌های مختلف نمودار قبلی را رسم کنید.</div>
<div dir='rtl'>پارامتر‌های <code>markersize</code>  و <code>linewidth</code> عدد طبیعی‌اند. همچنین در جدول زیر برخی از مقادیر ممکن برای پارامتر‌های <code>marker</code> و <code>linestyle</code> آمده است.</div>


| پارامتر    |مقدار |
|:----------|:------------------:|
| `marker`      |         '.'  '*'  '+' |
| `linestyle`       |          '-'  '--'  '-.'  ':'         |


````python
plt.plot(x, y, color='red', linewidth=1, linestyle=':', marker='+', markersize=10)
plt.show()
````

<div dir='rtl'>همان‌طور که احتمالا متوجه شدید با تعیین پارامتر <code>marker</code> می‌توان مکان دقیق داده‌ها را در نمودار نشان داد.</div>

````python
age = [21,12,32,45,37,18,28,52,5,40,48,15]
height = [160,135,170,165,173,168,175,159,105,171,155,158]
````

````python
# Set figure size
plt.figure(figsize=(12,6))

# Add a main title
plt.title("Plot of Age vs. Height (in cms)\n", fontsize=20, fontstyle='italic')

# X- and Y-label with fontsize
plt.xlabel("Age (years)", fontsize=16)
plt.ylabel("Height (cms)", fontsize=16)

# Turn on grid
plt.grid(True)

# Set Y-axis value limit
plt.ylim(100,200)

# X- and Y-axis ticks customization with fontsize and placement
plt.xticks([i*5 for i in range(12)], fontsize=15)
plt.yticks(fontsize=15)

# Main plotting function with choice of color, marker size, and marker edge color
plt.scatter(x=age, y=height, c='orange', s=150, edgecolors='k')

# Adding bit of text to the plot
plt.text(x=15, y=105, s="Height increaes up to around \n20 years and then tapers off", fontsize=15, 
         rotation=30, linespacing=2)
plt.text(x=22, y=185, s="Nobody has a height beyond 180 cm", fontsize=15)

# Adding a vertical line
plt.vlines(x=20, ymin=100, ymax=180, linestyles='dashed', color='blue', lw=3)

# Adding a horizontal line
plt.hlines(y=180, xmin=0, xmax=55, linestyles='dashed', color='red', lw=3)

# Adding a legend
plt.legend(['Height in cms'], loc=2, fontsize=14)

# Final show method
plt.show()
````

````python
mpg_csv = "mpg.csv"
````

````python
df = pd.read_csv(mpg_csv)
df.head()
````

<br>
<br>
<div dir='rtl'>به عنوان مثال نمودار پراکندگی مقدار مسافت طی‌شده با یک گالن سوخت (hwy) نسبت به اندازه موتور خودرو (displ) را رسم می‌کنیم.</div>


````python
plt.scatter(x=df['displ'], y=df['hwy'], c='blue', alpha=0.4)
plt.title('hwy - displ')
plt.xlabel('displ')
plt.ylabel('hwy')
plt.show()
````

<br>
<div dir='rtl'>برای استفاده از تابع <code>scatter</code> لازم است مقادیر هر بعد را مشخص نماییم. از طرفی با استفاده از پارامتر <code>alpha</code> میزان شفافیت هر نشان را مشخص می‌کنیم. بنابراین در مکان‌هایی که نشان پررنگ‌تر است، در واقع داده‌های بیشتری رو هم قرار گرفته‌اند.
برای تعیین عنوان نمودار از تابع <code>title</code> و برای نام‌گذاری محور‌ها از توابع <code>xlabel</code> و  <code>ylabel</code> استفاده می‌کنیم.</div>
<br>
<br>

````python
rng = np.random.RandomState(0)
colors = rng.rand(len(df))
sizes = 1000 * rng.rand(len(df))

#size متناسب با بزرگی اعداده
plt.scatter(x=df['displ'], y=df['hwy'], c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar()
````

<div dir='rtl'>حال می‌خواهیم با استفاده از ویژگی تعداد سیلندر داده‌ها را رنگ کنیم.</div>

````python
df.head()
````

````python
plt.scatter(x=df['displ'], y=df['hwy'], c=df['cyl'],alpha=0.5)
plt.title('hwy - displ')
plt.xlabel('displ')
plt.ylabel('hwy')
plt.colorbar()
plt.show()
````

<div dir='rtl'>به کمک تابع colorbar می‌توانیم طیف رنگ را نمایش دهیم. همان طور که انتظار داشتیم، خودرو‌های با تعداد سیلندر بیشتر مصرف سوخت بیشتری دارند.</div>

<div dir='rtl' id="barplot">
  <h2>
نمودار میله‌ای (bar plot)
     </h2>
</div>
<hr/>
<br>
<div dir='rtl'>با استفاده از این نمودار می‌توان کمیت عددی مربوط به مقادیر متفاوت یک کمیت دسته‌ای را نمایش داد و با هم مقایسه کرد.</div>

````python
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
scores = [20, 34, 30, 32, 27]
bar_width = 0.3
plt.bar(groups, scores, width=bar_width, color='black')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.show()
````

<br>
<div dir='rtl'>در نمودار میله‌ای بعد میانگین hwy و cty را برای کلاس‌های متفاوت خودرو‌ها به صورت همزمان نشان می‌دهیم.</div>

````python
df['class'].unique()
````

````python
classes = df['class'].unique()
barwidth = 0.3
cty_mean = []
hwy_mean = []
for x in classes:
    cty_mean.append(df[df['class'] == x]['cty'].mean())
    hwy_mean.append(df[df['class'] == x]['hwy'].mean())
````

````python
index = pd.factorize(classes)[0] + 1
index
````

````python
plt.bar(index, cty_mean, barwidth, color='black', label='mean city miles per gallon')
plt.legend()
plt.show()
````

````python
plt.bar(index, cty_mean, barwidth, color='black', label='mean city miles per gallon')
plt.legend()
plt.xticks(index, classes)
plt.show()
````

````python
plt.bar(index, hwy_mean, barwidth, color='red', label='mean highway miles per gallon')
plt.legend()
plt.xticks(index, classes)
plt.show()
````

````python
plt.bar(index, cty_mean, barwidth, color='black', label='mean city miles per gallon')
plt.bar(index + barwidth, hwy_mean, barwidth, color='purple', label='mean highway miles per gallon')
plt.xticks(index + barwidth / 2, classes)
plt.legend()
plt.show()
````

<div dir='rtl'>تابع <code>xticks</code> با گرفتن مکان دسته‌ها و نامشان، آن‌ها را روی محور x نمایش می دهد.
هنگامی که چند نمودار را در یک شکل رسم می‌کنیم، می‌‌توان با تعیین پارامتر <code>label</code> برای نمودار‌ها و فراخوانی تابع <code>legend</code>، برچسب تعیین شده برای هر نمودار را در شکل نشان دهیم.
دقت کنید در این شکل ما دو نمودار را به صورت همزمان به تصویر کشیدیم. در حالت کلی می‌توانیم تعداد دلخواهی نمودار با نوع مطلوب در یک شکل رسم کنیم.</div>

### Histogram

````python
weight = [55,35,77,68,70,60,72,69,18,65,82,48]
````
